## **Question 3**

You are optimizing a loss function $L(w) = w^4 - 10w^2 + 5w + 40$

Plot the loss function for the range $w ∈ [−3, 3]$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng()
space = np.arange(-3, 3, 0.01)
loss_function = lambda x: np.pow(x, 4) - (10 * np.pow(x, 2)) + (5 * x) + 40

# ----- Plotting -----
plt.figure(figsize=(10, 4))
plt.plot(space, loss_function(space), label='$L(w) = w^4 - 10w^2 + 5w + 40$')
# plt.ylim(0, 50)
plt.grid()
plt.legend(loc='upper right')
plt.show()
```
